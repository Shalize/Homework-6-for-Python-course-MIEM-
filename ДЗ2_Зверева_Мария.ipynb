```python
def sum_distance(from_num, to_num):
    # Меняем значения местами, если from_num больше to_num
    if from_num > to_num:
        from_num, to_num = to_num, from_num

    # Суммируем все числа от from_num до to_num включительно
    total = 0
    for num in range(from_num, to_num + 1):
        total += num

    return total
print(sum_distance(5, 1))
```

```python
def trim_and_repeat(string, offset=0, repetitions=1):
    trimmed_string = string[offset:]
    result = trimmed_string * repetitions
    return result
```

```python
# Запрашиваем ввод от пользователя
user_string = input("Введите строку: ")
user_offset = int(input("Введите число символов для обрезки (offset): ") or 0)
user_repetitions = int(input("Введите количество повторений (repetitions): ") or 1)
```

```python
# Вызываем функцию с пользовательскими параметрами
result = trim_and_repeat(user_string, user_offset, user_repetitions)
print("Результат:", result)
```
